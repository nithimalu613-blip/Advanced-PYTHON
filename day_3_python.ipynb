```python
def book_trip(transport,destination):  #keyword
  print(f"booking {transport} to {destination}")
book_trip(destination="paris",transport="flight")
```

```python
def book_trip(destination,transport):  #position
  print(f"booking {transport} to {destination}")
book_trip("london","train")
```

```python
def booking_trip(transport,destination="America"):  #default
  print(f"booking {transport} to {destination}")
booking_trip("flight")
```

```python
def booking_trip(destination,transport="plane"):
  print(f"booking {transport} to {destination}")
booking_trip("America","plane")
```

```python
def booking_trip(*destination,transport="plane"):   #arbitary
  print(f"booking {transport} to {destination}!")
booking_trip("America",transport="plane")
```
